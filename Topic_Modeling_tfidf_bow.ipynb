```python
def tokenize_text(text):
    #removing any parenthesis
    text = re.sub(r"\(\s*.*\s*\)", r'', text)
    text = re.sub(r"\[\s*.*\s*\]", r'', text)
    #remove characters except for [a-zA-Z0-9_] and single quotes assuming short forms are present like "won't"
    #which will be removed was stop words
    text = re.sub(r"[^\w\']", ' ', text)
    #converting text to lowercase to maintain uniformity across words
    text = text.lower()
    sent_tokenized = sent_tokenize(text)
    tokens = []
    for sent in sent_tokenized:
        tokens.extend(word_tokenize(sent))
    return tokens
```

```python
def remove_stopwords(words):
    stop_words = stopwords.words('english')
    stop_words.extend(['keywords', 'abstract'])
    words_without_stopwords = [word for word in words if word not in stop_words]
    return words_without_stopwords
```

```python
def preprocess(data):
    # Tokenizing and removing stop words in all docs
    print("no. of docs processing: {}".format(len(data)))

    words = []
    doc_no = 1
    for doc_text in data:
        print("Processing data in doc data_{}.txt".format(doc_no))
        #tokenizing text
        tokens = tokenize_text(doc_text)
        #removing stopwords
        tokens = remove_stopwords(tokens)
        #adding tokens to words
        words.append(tokens)
        #updating doc number
        doc_no += 1
    return words
```

```python
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
#preprocessed words
words = preprocess(read_file())

bigram = gensim.models.Phrases(words, min_count=3, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[words], threshold=100)  

# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)

# See trigram example
print(trigram_mod[bigram_mod[words]])

# def remove_stopwords(texts):
#     return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]

def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

```python
# data_words_nostops = remove_stopwords(words)

# Form Bigrams
data_words_bigrams = make_bigrams(words)

# Do lemmatization keeping only noun, adj, vb, adv
data_lemmatized = lemmatization(data_words_bigrams, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])

print(data_lemmatized)
```

```python
# importing tf-idf models from gensim
from gensim import models
```

```python
id2word = corpora.Dictionary(data_lemmatized)

# Create Corpus
texts = data_lemmatized

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# Term Freqency Inverse Document Frequency
tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]

# View
print(corpus[:])
```

```python
[[(id2word[id], freq) for id, freq in cp] for cp in corpus[:1]]
```

## Using bag of words as input to LDA

```python
#using bag of words as input to the model
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=7, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
print('\nPerplexity: ', lda_model.log_perplexity(corpus))  # a measure of how good the model is. lower the better.

# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, texts=data_lemmatized, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```
